```python
import pandas as pd
```

```python
Person=pd.read_excel("Person.xlsx",sheetname='Person',convertors={'SEX':str ,'OREC': str})
Diag=pd.read_excel("Person.xlsx",sheetname='Diag',convertors={'Diagnosis': str})

Person.sort_values(['HICNO'],inplace=True)
Diag.sort_values(['HICNO'],inplace=True)


cohort_study=pd.merge(Person[['HICNO','SEX','DOB']], Diag, how='inner', on='HICNO')
cohort_study.groupby(['HICNO','SEX','DOB']).count().reset_index().sort_values('Diagnosis',ascending=False)['Diagnosis']>0
cohort_study1=cohort_study[cohort_study.groupby(['HICNO','SEX','DOB']).count().reset_index().sort_values\
('Diagnosis',ascending=False)['Diagnosis']>0][['HICNO','SEX','DOB']]
cohort_study1.head()
# cohort_study = %sql \
#     SELECT d.PatientID, p.SEX, p.DOB \
#         from diagnoses d join mem.patients p\
#                 on d.PatientID = p.PATIENT_ID\
#         group by PatientID,p.SEX, p.DOB\
#         having COUNT(*) > 20\
#         order by Count(*) desc\
#         limit 100;
```

```python
def gender(l):
    if l == 'M':
        return 'male'
    else:
        return 'female'


cohort_patients = [Beneficiary(hicno=row[0], sex=gender(row[1]), dob=str(row[2].date()).replace("-","")) for index,row in cohort_study1.iterrows()]
```

```python
hicno = [pat.hicno for pat in cohort_patients]
hicno = tuple(hicno)
diags=Diag[Diag['HICNO']==hicno]
diags.head()
# diags = %sql SELECT DISTINCT PatientID, Diagnosis from diagnoses where PatientID in $hicno
diags = diags.set_index('HICNO')

for pat in cohort_patients:
    pat_diags = diags['Diagnosis'].loc[pat.hicno]
    pat.add_diagnosis(Diagnosis(pat,pat_diags,ICDType.TEN))
#     print(pat.diagnoses)
#     for code in pat_diags:
#         pat.add_diagnosis(Diagnosis(pat,code,ICDType.NINE))
```
